```python
Question-1

Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

       1

     /    \

  -2      3

  / \    /  \

4   5  -6   2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4   5

Also, entire tree sum is also 7.
```

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def maxSubtreeSum(root):
    global max_sum
    max_sum = float('-inf')

    def helper(node):
        global max_sum
        if node is None:
            return 0

        left_sum = helper(node.left)
        right_sum = helper(node.right)

        subtree_sum = node.val + left_sum + right_sum

        max_sum = max(max_sum, subtree_sum)

        return subtree_sum

    helper(root)

    return max_sum
```

```python
# Tree: Input1
#        1
#       / \
#      2   3
#     / \ / \
#    4  5 6  7

root1 = Node(1)
root1.left = Node(2)
root1.right = Node(3)
root1.left.left = Node(4)
root1.left.right = Node(5)
root1.right.left = Node(6)
root1.right.right = Node(7)

print(maxSubtreeSum(root1))
```

```python
# Tree: Input2
#        1
#       / \
#     -2   3
#     / \  / \
#    4   5 -6 2

root2 = Node(1)
root2.left = Node(-2)
root2.right = Node(3)
root2.left.left = Node(4)
root2.left.right = Node(5)
root2.right.left = Node(-6)
root2.right.right = Node(2)

print(maxSubtreeSum(root2))
```

```python
Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10
```

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def constructBST(level_order):
    if not level_order:
        return None

    root = Node(level_order[0])
    queue = [root]
    i = 1

    while i < len(level_order):
        current = queue.pop(0)
        left_value = level_order[i]
        right_value = level_order[i + 1] if i + 1 < len(level_order) else None

        if left_value is not None:
            current.left = Node(left_value)
            queue.append(current.left)

        if right_value is not None:
            current.right = Node(right_value)
            queue.append(current.right)

        i += 2

    return root
```

```python
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(level_order)

# Function to print the BST in inorder traversal
def inorderTraversal(node):
    if node is None:
        return

    inorderTraversal(node.left)
    print(node.val, end=" ")
    inorderTraversal(node.right)

# Print the constructed BST
inorderTraversal(root)
```

```python
Question-3

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.
```

```python
def can_represent_bst(arr):
    n = len(arr)
    if n == 0:
        return True

    # Initialize the root as the first element
    root = arr[0]

    # Create an empty stack and push the root onto the stack
    stack = []
    stack.append(root)

    # Initialize current minimum and maximum values
    curr_min = float('-inf')
    curr_max = float('inf')

    # Traverse the array from the second element onward
    for i in range(1, n):
        if len(stack) == 0:
            return False

        # Pop elements from the stack until the current element is less than the top element
        while len(stack) > 0 and arr[i] > stack[-1]:
            parent = stack.pop()

        # If the current element is outside the range, return False
        if arr[i] < curr_min or arr[i] > curr_max:
            return False

        # Update the current minimum and maximum values based on the parent (if it exists)
        if 'parent' in locals():
            curr_min = parent
        curr_max = parent if 'parent' in locals() else curr_max

        # Push the current element onto the stack
        stack.append(arr[i])

    return True

# Test cases
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
arr2 = [11, 6, 13, 5, 12, 10]

print(can_represent_bst(arr1))  # Output: Yes
print(can_represent_bst(arr2))  # Output: No
```
